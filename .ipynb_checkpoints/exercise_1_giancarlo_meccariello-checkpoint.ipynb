```python
import matplotlib
import pandas as pd
import wikipediaapi
```

```python
wiki = wikipediaapi.Wikipedia('en')
twitter = wiki.page('Category:Twitter')
youtube = wiki.page('Category:YouTube')
```

```python
twitter.text
```

```python
youtube.text
```

```python
def categories(name):
	mdict = {}
	for c in name.values():
		if not 'Category:' in c.title:
			print(f'_{c.title} \n')
			print(f'_{c.summary} \n\n')
			mdict[c.title] = c.summary
		else:
			print('Category: found')
	return mdict

```

```python
twitter_dict = categories(twitter.categorymembers)
```

```python
twitter_dict.keys()
```

```python
youtube_dict = categories(youtube.categorymembers)
```
